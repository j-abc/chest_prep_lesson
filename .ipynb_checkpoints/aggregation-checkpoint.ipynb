```python
from running import *
import os
import pandas as pd
```

```python
models = prepare_models_list()
```

```python
# build full model info including name, model params, and run params
full_model_info = []
for model, model_params, run_params in models:
    pre_model_name = choose_model(model, {**model_params,**{'compile_model':False}})[1]
    model_name      = prepare_model_name(pre_model_name,**run_params)
    full_model_info.append({'model_name': model_name, 'model': model, 'model_params': model_params, 'run_params':run_params})
```

```python
full_model_info
```

```python
# load model history
full_result_dir = '/home/jupyter/chest_models/'

for mdict in full_model_info:
    mdict['history_path'] = os.path.join(full_result_dir, mdict['model_name'], mdict['model_name'] + '_history.csv')
    mdict['last_model'] = os.path.join(full_result_dir, mdict['model_name'], mdict['model_name'] + '_end.h5')
    mdict['best_model'] = os.path.join(full_result_dir, mdict['model_name'], mdict['model_name'] + '.csv')
    mdict['first_model'] = os.path.join(full_result_dir, mdict['model_name'], mdict['model_name'] + '_first.h5')
    history = pd.read_csv(mdict['history_path'])
    history['model_name'] = mdict['model_name']
    history['epoch'] = history['epoch'] + 1
    mdict['history'] = history
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def plot_train_val(df, augment, dropout, learning_rate, num_layers):
    all_filter = ((df.augment == augment) & (df.dropout == dropout) & (df.learning_rate == learning_rate) & (df.num_layers == num_layers))
    history = df[all_filter]['history'].values[0]
    p = sns.lineplot(x = 'epoch', y = 'val_acc', data = history)
    return p

interact(lambda augment, dropout, learning_rate, num_layers: plot_train_val(cnn_df, augment, dropout, learning_rate, num_layers),
        augment = [('Off', 0), ('On',1)], 
        dropout = [('0%',0), ('30%',3), ('60%',6)], 
        learning_rate = [('1',0), ('0.001',3), ('0.000001',6)], 
        num_layers = [('1',1), ('2',2), ('4',4)])
```

```python
plot_train_val(cnn_df, 0, 0, 3, 1)
```
